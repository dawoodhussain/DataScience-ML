## formulae to find slope(m), intercept(b) using OLS Method:
### m = sum((x_train - x_mean) * (y_train - y_mean))/sum((x_train - x_mean)**2)
### b = y_mean - m*x_mean

## more details are mentioned in the below link:
### https://towardsdatascience.com/linear-regression-simplified-ordinary-least-square-vs-gradient-descent-48145de2cf76

```python
#load the test data from below link
# https://www.kaggle.com/andonians/random-linear-regression/data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df_train = pd.read_csv(r'E:\SimpliLearn\Data Science with Python\ML Algorithms Implementation\Linear Regression using Gradient\train.csv')
df_test = pd.read_csv(r'E:\SimpliLearn\Data Science with Python\ML Algorithms Implementation\Linear Regression using Gradient\test.csv')

#print(df_train.head(4))
#print(df_test.head(4))

#look at the shape of the data - (rows, columns)
print(df_train.shape, df_test.shape)

#check if any null or NAN values present in the data set
print(df_train.isnull().values.any())
print(df_test.isnull().values.any())

#remove the NAN values from dataset as observed above else the calculations would go wrong
df_train = df_train.dropna()

#from above set, let's get train & test X, Y values
x_train = df_train['x']
y_train = df_train['y']
x_test = df_test['x']
y_test = df_test['y']

#convert it to numpy array so that we can perform numerical computing on these arrays down the line
x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
y_test = np.array(y_test)
```

```python
#Let's plot the graph against train dataset to see how data is spread
plt.figure(figsize=(12,12))
plt.scatter(x_train,y_train)
plt.show()
```

## Finding the best fit line for above datapoints

```python
#get the values (m,b) using OLS formulae
x_mean = x_train.mean()
y_mean = y_train.mean()
m = np.sum((x_train - x_mean)*(y_train - y_mean))/np.sum((x_train - x_mean)**2)
b = y_mean - (m*x_mean)

print(m,b)
```

```python
y_train_predict = (m*x_train) + b
residual_err = np.sum(y_train_predict - y_train)
residual_err
```

```python
#determining coefficient of determination
from sklearn.metrics import r2_score
y_test_predict = (m*x_test) + b
print("R2 Score : ",r2_score(y_test, y_test_predict))
```
